```python
ticker = 'APPL' 

start_date = str(int(datetime.datetime(2021,10,1).timestamp()))
end_date = str(int(datetime.datetime(2021,10,27).timestamp())) 

interval= '1d'
#'1wk'
#'1mo'
#'1yr'

event ='history'

url = 'https://query1.finance.yahoo.com/v7/finance/download/'\
      + ticker + '?period1=' + start_date + '&period2=' + end_date + '&interval='\
      + interval + '&events=' + event 

#'&includeAdjustedClose=true'

print(url)
print (start_date, end_date)
```
